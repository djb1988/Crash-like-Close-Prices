```python
# prompt: The relative maximum of the time t is defined as:
# St = max( Yu , u ≤ t )
#  for each point of the time series, the DrawDown is computed as the difference between the relative maximum and the time series:
# DD_t = St – Yt. compute maximum of drawdown  of the dataframe df of close values if there are 50 epochs

def max_drawdown(df):
  """
  Calculates the maximum drawdown of a given dataframe of close values.

  Args:
      df: A pandas dataframe containing close values.

  Returns:
      The maximum drawdown as a float.
  """

  # Initialize variables
  max_value = df['Close'].max()
  max_drawdown = 0

  # Loop through each row in the dataframe
  for i in range(len(df)):
    # Calculate the current drawdown
    current_drawdown = max_value - df.iloc[i]['Close']

    # Update the maximum drawdown if the current drawdown is greater
    if current_drawdown > max_drawdown:
      max_drawdown = current_drawdown

  # Return the maximum drawdown
  return max_drawdown

# Calculate the maximum drawdown for the dataframe
max_drawdown = max_drawdown(df)

# Print the maximum drawdown
print(f"Maximum drawdown: {max_drawdown}")

```

```python


from scipy.stats import norm
import numpy as np
import pandas as pd
import sqlite3 as sql


samples = range(1, 2048 + 1)
epoch = range(1, 20000 + 1)
DATASET = pd.DataFrame(index = samples, columns = ['media', 'devst', 'periodo'])
TARGET  = pd.DataFrame(index = samples, columns = ['MDD'])
#Nested loop, in the outer one the value of mean, standard deviation and period of the epoch are randomly defined within fixed intervals, in the inner one the equity and the 90th percentile of the MDD are calculated:

for j in samples :
    Risultato = np.zeros( ( len(epoch) ) )
    media     = np.random.uniform(low = 0.00005, high = 0.00035)
    devst     = np.random.uniform(low = 0.0005, high = 0.0150)
    periodo   = np.random.randint(8, high = 32, size = None, dtype = int)
    for i in epoch :
        LogReturn = norm.rvs(loc = media, scale = devst, size = periodo)
        cumulato = np.cumsum(LogReturn)
        Equity = 1000 * np.exp(cumulato)
        DD_rif = np.maximum.accumulate(Equity)
        DrawDown = (DD_rif - Equity) / DD_rif
        Risultato[i-1] = DrawDown.max()

    DATASET.loc[j, 'media'] = media
    DATASET.loc[j, 'devst'] = devst
    DATASET.loc[j, 'periodo'] = periodo
    TARGET.loc[j, 'MDD'] = np.percentile(Risultato, 90)
import sqlite3

conn = sqlite3.connect('Synthetic.sqlite')
cur = conn.cursor()
cur.execute('''
CREATE TABLE IF NOT EXISTS dataset (
    "indice" INTEGER NOT NULL UNIQUE,
    "media" REAL,
    "devst" REAL,
    "periodo" REAL,
    "MDD" REAL,
    PRIMARY KEY ("indice" AUTOINCREMENT)
)
''')

#cur.execute('CREATE TABLE IF NOT EXISTS dataset (')
   # "indice" INTEGER NOT NULL UNIQUE,
    #"media" REAL,
    #"devst" REAL,
    #"periodo" REAL,
    #"MDD" REAL,
    #PRIMARY KEY ("indice" AUTOINCREMENT) );' )

for i in range(DATASET.shape[0]):
    cur.execute('''
    INSERT INTO dataset (media, devst, periodo, MDD)
    VALUES (?, ?, ?, ?)
    ''',
    (DATASET.iloc[i, 0],
     DATASET.iloc[i, 1],
     int(DATASET.iloc[i, 2]),
     TARGET.iloc[i, 0]))
    conn.commit()
cur.close()
```

```python
# prompt: using isometric curves, plot standard deviation on x-axis and period on y-axis with  90 MDD given an arbitrary coefficient

# Define the coefficient
coefficient = 0.5

# Generate random data for standard deviation and period
std_devs = np.linspace(0.0005, 0.0150, 100)
periods = np.linspace(8, 32, 100)

# Calculate the 90MDD for each pair of standard deviation and period
mdd_90 = coefficient * std_devs + periods

# Create the plot
fig, ax = plt.subplots()

# Plot the isometric curves
for mdd in np.arange(0.05, 0.25, 0.01):
  ax.plot(std_devs, (mdd - coefficient * std_devs) / periods, color='gray', alpha=0.5)

# Plot the scatter plot
ax.scatter(std_devs, periods, c=mdd_90, cmap='viridis')

# Add colorbar
cbar = fig.colorbar(ax.collections[0])
cbar.set_label('90MDD')

# Set axis labels
ax.set_xlabel('Standard Deviation')
ax.set_ylabel('Period')

# Add title
ax.set_title('90MDD as a Function of Standard Deviation and Period')

# Show the plot
plt.show()
```
